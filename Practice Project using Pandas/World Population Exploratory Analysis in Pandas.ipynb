```python
# Group by Continent and find the country with the highest population in 2022 for each continent

    # .apply() says: "For each group, run the function I provide."

    # x.loc[...], returns the entire row of every x found

    # .idxmax() is a Pandas method that returns the index label of the maximum value.
    # .max() returns the maximum value while .idxmax() returns the location (index) of the maximum value.

highest_population_2022 = df.groupby('Continent').apply(lambda x: x.loc[x['2022 Population'].idxmax()])
highest_population_2022
```

```python
# finding country with highest growth rate

df.sort_values(by='Growth Rate', ascending=False)
```

```python
# Understanding share of global population

df['World Population Percentage'].describe()
```

```python
df.groupby('Continent').mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)
```

```python
# grouping by continent
# using reverse, the columns be in a chronological order

    # [::-1], means:
                    # Start at the end
                    # Move backward one step at a time

    # df.columns[5:13][::-1], means: "Take columns 5 through 12, then reverse their order."

df2 = df.groupby('Continent')[df.columns[5:13][::-1]].mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)
df2
```

```python
# a[2:8]     - positions 2 to 7
# a[::-1]    - reverse everything
# a[::2]     - every second item
# a[1::2]    - every second item starting at index 1
# a[::-2]    - reverse, skipping every other item

# EXAMPLE

a = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']

                            # a[start, stop, step]

print(a[::2])   # Means: Start at the beginning, Go to the end, Take every 2nd item

print(a[1::2])   # Means: Start at index 1, Go to the end, Take every 2nd item

print(a[::-1])   # Means: Start from the end, Move backwards one step at a time

print(a[::-2])   # Means: Start from the end, Move backwards by 2 positions each time
```
